## Product Analytics - Data Dictionnary

### Events (`./datasets/product/instance_events.csv`)

_User events (results of actions) tracked in Instances_

__Columns__:

- `event_date`: Date when the event occurs
- `event_id`: Unique identifier of the event
- `event_name`: Name of the event tracked
    - `Instance Login`: User log in to an instance
    - `Small App Opened`: User open a small app
    - `Story Opened`: User open a story
    - `Home Opened`: User open home
    - `Live query created`: User create live queries
    - `Connector configuration`: User open connector configuration
    - `Dataset configuration`: User open dataset configuration
    - `Prepared dataset created`: User create prepare dataset
    - `SQL Query Editor opened`: User open SQL Query Editor
    - `Logout`: User log out from an instance
    - `Home preview`: User preview home
    - `Glossary tab opened`: User open glossary
    - `App history opened`: User open app changes history
    - `Create Blank App`: User create new blank small app
    - `Notifications`: User set up notifications
    - `Restore app`: User restore deleted small app
    - `Settings`: User access to instance settings options
- `device`: Device used when user performs action
- `browser`: Browser used when user performs action
- `smallappname`: Name of the small
- `user_uuid`: Unique identifier of the user
- `instance_uuid`: Unique identifier of the instance


### Users (`./datasets/sales/instance_users.csv`)

_List of users registered in Instances_

__Columns__:

- `user_fullname`: Fullname of the user
- `user_email`: Email of the user
- `user_uuid`: Unique identifier of the user 
- `user_country`: Last country the user connected from
- `user_dept`: Department where the user is working 


### Events classification (`./datasets/product/instance_event_classification.csv`)

_Event classification per user type_ 

- `event_name`: Name of the event tracked
- `is_app_builder_key_event`: Boolean flag to identify events that most corresponding to App Builders (key events). App Builders are the one who produce the content (stories and chart)  
- `is_admin_key_event`: Boolean flag to identify events that most corresponding to Admin (key events). Admin are the instance administrators who are responsible of security (user and small app management)
- `is_key_viewer_event`: Boolean flag to identify events that most corresponding to Viewers (key events). Viewers are the end users who are going to consume/read the content.
